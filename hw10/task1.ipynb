````python
import pandas as pd
import random
lst = ['robot'] * 10
lst += ['human'] * 10
print(random.shuffle(lst))
data = pd.DataFrame({'whoAmI':lst})
data
````

````python
def one_hot_func(data_frame, col):
    first_value = list(set(data_frame[col]))[0]
    second_value = list(set(data_frame[col]))[1]
    data_frame.loc[data_frame[col] == first_value, col] = 1
    data_frame.loc[data_frame[col] == second_value, col] = 0
one_hot_func(data,'whoAmI')
data
````
